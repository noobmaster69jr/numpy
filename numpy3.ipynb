```python
# Stacking
```

```python
data = np.arange(5)
data
```

```python
np.vstack([data, data, data])
```

```python
data = data.reshape(5, 1)
data
```

```python
np.hstack([data, data, data])
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.hstack((a, b)))
```

```python
a = np.array([[1], [2], [3]])
b = np.array([[4], [5], [6]])
np.hstack((a, b))
```

```python
# concatenate
```

```python
z = np.array([[2, 4]])
z
```

```python
np.concatenate([z, z], axis=0)
```

```python
np.concatenate([z, z], axis=1)
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6]])
np.concatenate((a, b), axis=0)
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6]])
np.concatenate((a, b.T), axis=1)
```
